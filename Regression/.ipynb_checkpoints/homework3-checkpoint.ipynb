### Проверка зависимости от размера выборки

```python
coefs = dict(
    b1 = [],
    b1_down = [],
    b1_up = [],

    b2 = [],
    b2_down = [],
    b2_up = [],
)
for i in tqdm(range(30, 1001)):
    dist_params['N'] = i
    data = gen_data(y_type="multivariate", params=dist_params, seed=RANDOM_SEED)
    _, model = train_model(data, target='y', feature_names=['x1', 'x2'], pairwise=True)

    interval = model.conf_int()

    # print(model.params)
    # print(interval)

    coefs['b1'].append(model.params[1])
    coefs['b1_down'].append(interval.iloc[1][0])
    coefs['b1_up'].append(interval.iloc[1][1])

    coefs['b2'].append(model.params[2])
    coefs['b2_down'].append(interval.iloc[2][0])
    coefs['b2_up'].append(interval.iloc[2][1])
```

#### Коэффициент $\beta_1$

```python
plt.style.use('ggplot')
plt.figure(figsize=(13,8))
X = [i for i in range(0, 971)]

plt.fill_between(X, coefs['b1_down'], coefs['b1_up'], color='#98FB98', label='Доверительный интервал')
plt.plot(coefs['b1'], label='beta 1')
plt.axhline(dist_params['beta1'], color='b', linestyle='dashed', label='beta 1 true')
plt.title('Оценки коэффициентов и их доверительные интервалы')
plt.xlabel('Размер выборки')
plt.ylabel('Оценки коэффициентов')
plt.legend()
plt.show()
```

#### Коэффициент $\beta_2$

```python
plt.style.use('ggplot')
plt.figure(figsize=(13,8))
X = [i for i in range(0, 971)]

plt.fill_between(X, coefs['b2_down'], coefs['b2_up'], color='#98FB98', label='Доверительный интервал')
plt.plot(coefs['b2'], label='beta 2')
plt.axhline(dist_params['beta2'], color='b', linestyle='dashed', label='beta 2 true')
plt.title('Оценки коэффициентов и их доверительные интервалы')
plt.xlabel('Размер выборки')
plt.ylabel('Оценки коэффициентов')
plt.legend()
plt.show()
```

### Вывод
При увелечении размера выборки длина доверительного интервала уменьшается, и разброс оценочных коэффициентов также сходится к истинному значению.
